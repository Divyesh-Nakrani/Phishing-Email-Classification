```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("dataset_HP.csv")
df.isnull().sum()
```

```python
df_x=df.dropna()
```

```python
key=np.arange(0,len(df_x.index))

df_x.set_index(keys=key,inplace=True)
df_x.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df_x["label"]=le.fit_transform(df_x["label"])
df_x.head()

# for ham zero and for phish its 1
```

# Splitting Train & Test Data

```python
x=df_x.iloc[:,:-1]
y=df_x.iloc[:,-1]

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
print("shape of x_train is",x_train.shape)
print("shape of x_test is",x_test.shape)
print("shape of y_train is",y_train.shape)
print("shape of y_test is",y_test.shape)
```

# Try with k-fold cross validation

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


model_tr=DecisionTreeClassifier()
model_rd=RandomForestClassifier()
model_sv=SVC()

kf=KFold(n_splits=5)
acc_tr=cross_val_score(model_tr,x,y,cv=kf)
print("--> Decision tree Classifier accuracy using k fold is ",acc_tr ,"\nAverage of all is",acc_tr.mean())
print("\n")
acc_rd=cross_val_score(model_rd,x,y,cv=kf)
print("--> Random forest Classifier accuracy using k fold is ",acc_rd ,"\nAverage of all is",acc_rd.mean())
print("\n")
acc_sv=cross_val_score(model_sv,x,y,cv=kf)
print("--> static vector Classifier accuracy using k fold is ",acc_sv ,"\nAverage of all is",acc_sv.mean())
```

# Trained model with hyperparameter tuning

1.started with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score

model_lr= LogisticRegression()
model_lr.fit(x_train,y_train)
pred_y=model_lr.predict(x_test)
print("Accuracy using LogisticRegression ",accuracy_score(y_test,pred_y))
print("\n")
print("classification report using Logistic regression is \n",classification_report(y_test,pred_y))
```

2.Using Random forest classfier

```python
from sklearn.ensemble import RandomForestClassifier

model_rf= RandomForestClassifier(n_estimators=1000)
model_rf.fit(x_train,y_train)
pred_y=model_rf.predict(x_test)
print("Accuracy using Random Forest Classifier ",accuracy_score(y_test,pred_y))
print("\n")
print("classification report using Random Forest Classifier is \n",classification_report(y_test,pred_y))
```

3.Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model_tr= DecisionTreeClassifier()
model_tr.fit(x_train,y_train)
pred_y=model_tr.predict(x_test)
print("Accuracy using Decision Tree Classifier ",accuracy_score(y_test,pred_y))
print("\n")
print("classification report using Decision Tree Classifier is \n",classification_report(y_test,pred_y))
```

4.Using K-Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

model_kn= KNeighborsClassifier()
model_kn.fit(x_train,y_train)
pred_y=model_kn.predict(x_test)
print("Accuracy using K Neighbors Classifier ",accuracy_score(y_test,pred_y))
print("\n")
print("classification report using K Neighbors Classifier is \n",classification_report(y_test,pred_y))
```
